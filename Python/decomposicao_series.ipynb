```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
base = pd.read_csv('dados/AirPassengers.csv')
#fazendo parse da data
dateparse = lambda dates:pd.datetime.strptime(dates, '%Y-%m')
base = pd.read_csv('dados/AirPassengers.csv', parse_dates = ['Month'], index_col = 'Month', date_parser=dateparse)
base
```

```python
#transformando dataframe para series
ts = base['#Passengers']
```

```python
plt.plot(ts)
```

```python
decomposicao = seasonal_decompose(ts)
tendencia = decomposicao.trend
sazonal = decomposicao.seasonal
aleatorio = decomposicao.resid
```

```python
plt.plot(tendencia)
```

```python
plt.plot(sazonal)
```

```python
plt.plot(aleatorio)
```

```python
plt.subplot(4,1,1)
plt.plot(ts, label='Original')
plt.legend(loc='best')

plt.subplot(4,1,2)
plt.plot(tendencia, label='TendÃªncia')
plt.legend(loc='best')

plt.subplot(4,1,3)
plt.plot(sazonal, label='Sazonalidade')
plt.legend(loc='best')

plt.subplot(4,1,4)
plt.plot(aleatorio, label='Aleatoriedade')
plt.legend(loc='best')

plt.tight_layout()
```
